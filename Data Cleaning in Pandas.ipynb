```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Data Analyst\Python\Pandas\Customer Call List.xlsx")
```

```python
df
```

```python
#Deleting Duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
#Fixing Firstname and lastname
df['Last_Name'] = df['Last_Name'].str.lstrip()
df['Last_Name'] = df['Last_Name'].str.strip(".../_")
```

```python
df['Last_Name'] = df['Last_Name'].apply(lambda x : str(x))
```

```python
#Fixing Phone Number
df['Phone_Number'] = df['Phone_Number'].str.replace("[^A-Za-z0-9]","")
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--','')
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('Na--','')
```

```python
df
```

```python
#spliting the address
df[['Street Address','State','Postal Code']] = df['Address'].str.split(',',2,expand=True)
```

```python
df
```

```python
#deleting Useless columns
df = df.drop(columns = "Not_Useful_Column")
```

```python
#fixing Paying Customer and Do_Not_Contact
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].replace('No','N')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('No','N')
```

```python
df
```

```python
#Showing only those which we can contact them
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace = True)
df
```

```python
#Removing null values in Phone number column
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace = True)
df
```

```python
#replacing none with blank
df = df.fillna('')
```

```python
#Removing Null Values
df['Last_Name'] = df['Last_Name'].replace('nan','')
```
